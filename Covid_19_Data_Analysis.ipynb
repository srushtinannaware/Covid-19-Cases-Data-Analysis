<h1>Covid-19 Data Analysis Using Pandas and Data Visualiazation using Matplotlib</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Covid-19 Daily Cases</h2>
<p>Data Visualization Using Simple Plot</p>

```python
data = pd.read_csv('case_time_series.csv')
```

```python
data.head()
```

```python
Y = data.iloc[61:,1].values
R = data.iloc[61:,3].values
D = data.iloc[61:,5].values
X = data.iloc[61:,0]
```

```python
plt.figure(figsize=(30,8))
plt.plot(X,Y)
```

```python
plt.figure(figsize=(30,8))
plt.plot(X,Y)
plt.xticks(rotation=40,size='20',color='black')
plt.yticks(size=20,color='black')
plt.tick_params(size=20,color='black')
ax = plt.axes()
ax.grid(linewidth=0.4,color='black')
ax.set_xlabel('Date',color='black')
ax.set_ylabel('Number of Confirmed Cases',color='black')
ax.plot(X,Y,color='blue',marker='o',linewidth=4,markersize=15)
```

```python
plt.figure(figsize=(30,8))
plt.plot(X,Y)
plt.xticks(rotation=40,size='20',color='black')
plt.yticks(size=20,color='black')
plt.tick_params(size=20,color='black')
ax = plt.axes()
ax.grid(linewidth=0.4,color='black')
ax.set_xlabel('Date',color='black')
ax.set_ylabel('Number of Confirmed Cases',color='black')
ax.plot(X,Y,color='blue',marker='o',linewidth=4,markersize=15)

for i,j in zip(X,Y):
    ax.annotate(str(j),xy=(i,j+100),color='black',size='13')
     
ax.annotate('Second Lockdown 15th April',
            xy=(15.2, 860),
            xytext=(19.9,500),
            color='black',
            size='25',
            arrowprops=dict(color='black',
                            linewidth=0.025))
 
plt.title("COVID-19 IN : Daily Confirmed\n",
          size=50,color='#28a9ff')
 
ax.plot(X,Y,
        color='#1F77B4',
        marker='o',
        linewidth=4,
        markersize=15,
        markeredgecolor='#035E9B')
```

<h2>How the virus is spreading based on Travel, Place Visit,etc.</h2>
<p>Data Visualization using Pie Chart</p>

```python
slices = [62, 142, 195]
activities = ['Travel', 'Place Visit', 'Unknown']
 
cols=['blue','yellow','red']
exp = [0.2,0.02,0.02]
 
plt.pie(slices,labels=activities,
        textprops=dict(size=25,color='black'),
        radius=3,
        colors=cols,
        autopct='%2.2f%%',
        explode=exp,
        shadow=True,
        startangle=90)
 
plt.title('Transmission\n\n\n\n',color='#4fb4f2',size=40)
```

<h2>The district wise coronavirus cases for a state.</h2>
<p>Data Visualization using Bar Plot</p>

```python
data = pd.read_csv('district.csv')
data.head()
```

```python
re=data.iloc[:30,5].values
de=data.iloc[:30,4].values
co=data.iloc[:30,3].values
x=list(data.iloc[:30,0])
```

```python

plt.figure(figsize=(25,10))
ax=plt.axes()
 
ax.grid(linewidth=0.4, color='black')
 
 
plt.xticks(rotation=60,
           size='20',
           color='black')#ticks of X
 
plt.yticks(size='20',color='black')
 
 
ax.set_xlabel('\nDistrict',size=25,
              color='black')
ax.set_ylabel('No. of cases\n',size=25,
              color='black')
 
 
plt.tick_params(size=20,color='black')
 
 
ax.set_title('Maharashtra District wise breakdown\n',
             size=50,color='#28a9ff')
 
plt.bar(x,co,label='re')
plt.bar(x,re,label='re',color='green')
plt.bar(x,de,label='re',color='red')
 
for i,j in zip(x,co):
    ax.annotate(str(int(j)),
                xy=(i,j+3),
                color='black',
                size='15')
 
plt.legend(['Confirmed','Recovered','Deceased'],
           fontsize=20)
```
